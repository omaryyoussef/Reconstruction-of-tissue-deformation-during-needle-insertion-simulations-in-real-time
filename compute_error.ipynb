```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
true_train = np.load('truetrain.npy')
out_train = np.load('outtrain.npy')
true_valid = np.load('truevalid.npy')
out_valid = np.load('outvalid.npy')
```

```python
# Training Error
groundtruth = true_train
finalO = out_train

error_train = abs(np.subtract(groundtruth, finalO))
error_train = error_train.flatten()
error_train = error_train[error_train > 0.1] # 0.1 to look at large errors

print('Shape of training error array: ', np.shape(error_train))
print('Maximum training error value: ', max(error_train))
print('Minimum training error value: ', min(error_train))
print('Mean training error value: ', np.mean(error_train))
print('STD training error value: ', np.std(error_train))
```

```python
#hist_train, bin_edges_train = np.histogram(error_train)
plt.hist(error_train, bins = 'auto', color='#0504aa', alpha=0.7)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Training Error Values > 0.1')
plt.ylabel('Frequency') # count
plt.title('Training Error Histogram')
plt.show()
```

```python
# Validation Error
groundtruth_valid = true_valid
finalO_valid = out_valid

error_valid = abs(np.subtract(groundtruth_valid, finalO_valid))
error_valid = error_valid.flatten()
error_valid = error_valid[error_valid > 0.1] # 0.1 to look at large errors

print('Shape of validation error array: ', np.shape(error_valid))
print('Maximum validation error value: ', max(error_valid))
print('Minimum validation error value: ', min(error_valid))
print('Mean validation error value: ', np.mean(error_valid))
print('STD validation error value: ', np.std(error_valid))
```

```python
#hist_valid, bin_edges_valid = np.histogram(error_valid, bins = 'auto')
plt.hist(error_valid, bins = 'auto', color='#0504aa', alpha=0.7)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Validation Error Values > 0.1')
plt.ylabel('Frequency')
plt.title('Validation Error Histogram')
plt.show()
```
